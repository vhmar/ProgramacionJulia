```julia
using PyPlot
```

```julia
x=collect(0:200);
t=collect(0:200);
```

```julia
k=4.56 ;  #long de onda
w=3.09 ; #frecuencia
```

```julia
plot(sin.(k*x).*cos.(w*t),"r")
plot((sin.(k*x+w*t))/2 + (sin.(k*x-w*t))/2, "b")
```

```julia
plot(t, sin.(k*x).*cos.(w*t),"r")
plot(t, (sin.(k*x+w*t))/2 + (sin.(k*x-w*t))/2, "b")
```
