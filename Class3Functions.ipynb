```julia
function regre(x,y)
    #Regre Lineal
    n = length(x)
    sx = sum(x)
    sxs = sum(x.^2)
    sy = sum(y)
    sxy = sum(x.*y)
    m = n*sxs-sx^2
    a = (sy*sxs - sx*sxy)/m
    b = (n*sxy - sx*sy)/m
    rms = sqrt(sum((y-a-b*x).^2)/n)
    return a,b,rms
end
```

```julia
a,b,rms = regre(x,yru)
```
