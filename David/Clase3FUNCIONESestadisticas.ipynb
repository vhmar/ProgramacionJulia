```julia
x1 = collect(1:1:50);
```

```julia
y = 2*x1 + 5;
rui = randn(length(y))*4
yrui = y + rui;
```

```julia
plot(x1,yrui,"o")
xlabel("X")
ylabel("Y")
```

```julia
a,b = linreg(x1,yrui)
```

```julia
yes=a + b*x1;
```

```julia
plot(x1,yrui,"ro")
plot(x1,yes,"k--")
plot(x1,y,"b-")
errorbar(x1,yes,rms)
```

```julia
rms = sqrt(sum((yrui-yes).^2)/length(yrui)) #rms es el error cuadratico medio
```

```julia
function compa(x,y)
    
    n=length(x)
    delta= n*sum(x.^2)-((sum(x))^2)
    a=((sum(x.^2)*sum(y))-(sum(x)*sum(x.*y)))/delta
    b=(n*sum(x.*y)-(sum(x)*sum(y)))/delta
    rms=sqrt(sum((y-a-b*x).^2)/n)
    return a,b,rms
end
```

```julia
a,b,rms = compa(x1,yrui)
```
